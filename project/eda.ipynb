```python
df.drop(['company','agent'],axis=1,inplace=True)
```

```python
df=df[df['adr']<5000] #to remove outlier
```

```python
#data analysis
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

#plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Resevation status in different hotels',size=20)
plt.legend(['not canceled','canceled'])
plt.xlabel('hotel')
plt.ylabel('number of reservations')
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python

# 1) build counts and convert to percentages within each hotel
counts = df.groupby(['hotel', 'is_canceled']).size().unstack(fill_value=0)
perc = counts.div(counts.sum(axis=1), axis=0) * 100   # row-wise percentage

# 2) plot grouped bars of percentages
ax = perc.plot(kind='bar', figsize=(8,5))
ax.set_ylabel('Percentage (%)')
ax.set_xlabel('Hotel')
ax.set_title('Cancellation percentage within each hotel')

# 3) annotate percent labels on top of bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='edge')

# 4) make legend text readable
ax.legend(['Not canceled (0)', 'Canceled (1)'], bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel')
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
```

```python
import calendar

df['month_num'] = df['reservation_status_date'].dt.month
df['month'] = df['reservation_status_date'].dt.strftime('%b')  # Jan, Feb, ...

df.sort_values('month_num',inplace=True)
plt.figure(figsize=(16,8))
plt.title('Monthly analysis of cancelation and non-cancelation')
ax=sns.countplot(x='month',hue='is_canceled',data=df)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge')

ax.legend(['Not canceled (0)', 'Canceled (1)'], bbox_to_anchor=(1,1))
plt.show()
```

```python
plt.title('ADR per month')
adr_per_month = df[df['is_canceled']==1].groupby(['month', 'month_num'])[['adr']].sum().reset_index()
adr_per_month = adr_per_month.sort_values('month_num')
month_order = adr_per_month['month'].tolist()
sns.barplot(x='month',y='adr',data=adr_per_month,order=month_order)
plt.show()
```

```python
canceled_data=df[df['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```
